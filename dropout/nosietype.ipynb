## 1. 基于entanglement

```python
p = [-4, -3, -2]
acc_b = [96.49, 95.84, 92.21]
acc_b_1 = [98.70, 98.57, 96.91]
acc_b_2 = [98.76, 98.70, 95.48]
acc_bp = [98.16, 97.64, 95.37]
acc_bp_1 = [98.53, 98.22, 96.48]
acc_bp_2 = [98.74, 98.26, 96.25]
xlabels = ["0.0001", "0.001", "0.01"]
```

```python
plt.title("不同噪声类型下噪声概率与 QLSTM 准确率", fontsize=13)
plt.ylabel("accuarcy(%)", fontsize=13)
plt.xlabel("噪声概率(1-p)", fontsize=13)
plt.xticks(p, xlabels)
plt.plot(p, acc_b, 'o--', label="bit flip", color="#FF6666")
plt.plot(p, acc_b_1, 'o-', label="bit flip(dp=0.0001)", color="#FF99CC")
plt.plot(p, acc_b_2, 'o-', label="bit flip(dp=0.0002)", color="#EE3399")

plt.plot(p, acc_bp, 's--', label="bit-phase flip", color="#66CCFF")
plt.plot(p, acc_bp_1, 's-', label="bit-phase flip(dp=0.0001)", color="#00CCBB")
plt.plot(p, acc_bp_2, 's-', label="bit-phase flip(dp=0.0002)", color="#339999")
plt.legend()
```
