## 1.2 Transfer data to LSTM representation

```python
from sklearn.preprocessing import StandardScaler
```

```python
def data_process(data, window_size, predict_size):
    scaler = StandardScaler()
    data = scaler.fit_transform(np.array(data).reshape(-1, 1))
    
    data_in = []
    data_out = []
    
    for i in range(data.shape[0] - window_size - predict_size):
        data_in.append(data[i:i + window_size].reshape(1, window_size)[0])
        data_out.append(data[i + window_size:i + window_size + predict_size].reshape(1, predict_size)[0])
        
    data_in = np.array(data_in).reshape(-1, window_size)
    data_out = np.array(data_out).reshape(-1, predict_size)
    
    data_process = {'datain': data_in, 'dataout': data_out}
    
    return data_process, scaler
```

## 1.3 prepare train/test dataset

```python
features_size = 4
window_size = features_size * 3 # features num * time steps
predict_size = features_size # features
```

```python
test_processed, test_scaler = data_process(test, window_size, predict_size)
X_test, y_test = test_processed['datain'], test_processed['dataout']
```

```python
X_test = torch.from_numpy(X_test.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
```

# 2. Plot Training losses

```python
def plotLosses(losses, color="red", fill_color="#CCCCFF", title=None, fontsize=15):
    losses_mean = np.mean(losses, axis=0)
    losses_std = np.std(losses, axis=0)
    
    x = range(1, len(losses_mean) + 1)
    plt.xlabel("轮次", fontsize=13)
    plt.ylabel("损失", fontsize=13)
    plt.grid()
    plt.plot(x, losses_mean, color=color, linewidth=2.0, linestyle='--', label="平均值")
    plt.fill_between(x, losses_mean + losses_std, losses_mean - losses_std, 
                     color=fill_color, label="标准差")
    plt.legend(["平均值", "标准差"], loc="upper right")
    
    if title != None:
        plt.title(title, fontsize=fontsize)
```

```python
mode = 'entanglement'
```

# 3. Plot Results

## 3.1 load qmodel

```python
# maximum qubits size
ctx = InitQMachine(5)
```

```python
features_size = 4
window_size = features_size * 3
predict_size = features_size # features

input_size = window_size
num_output = predict_size

hidden_size = 32
num_layers = 2

# loaded quantum model
qmodel = QModel(input_size, hidden_size, num_output, 
                num_layers = num_layers, ctx = ctx, mode='GRU')
```

```python
qmodel.load_state_dict(torch.load(f"./model/{mode}/model_{mode}_5.pt"))
```

## 3.2 prepare results

```python
results = []

for i in range(0, len(X_test), features_size):
    input_data = X_test[i:i+1]
    result = qmodel(input_data)
    result = test_scaler.inverse_transform(result.data)
    
    results.append(result[0])
```

## 3.3 Plot features

```python
plt.figure(figsize=(20, 7))
plt.plot(test['meantemp'])
meantemp = [ r[0] for r in results ] 
meantemp = [None, None, None, None] + meantemp
plt.plot(meantemp, color="red")
plt.show()
```
