# 1. Circuits Sampler

## 1.1 Circuit A

```python
def sampleCircuitA(layer=1, qubits=4, encode=False):
    circuit = QuantumCircuit(qubits)
    
    if encode:
        for i in range(qubits):
            circuit.h(i)
            circuit.ry(np.random.uniform(0, 2 * np.pi), i)
        circuit.barrier()

    for i in range(layer):
        for i in range(qubits - 1):
            circuit.cx(i, i + 1)
        circuit.cx(qubits - 1, 0)
        
        for i in range(qubits - 1):
            circuit.cx(i, i + 1)
        circuit.cx(qubits - 1, 0)
        
        circuit.barrier()
        
        for i in range(qubits):
            circuit.u3(np.random.uniform(0, 2 * np.pi), np.random.uniform(0, 2 * np.pi), np.random.uniform(0, 2 * np.pi), i)

        circuit.barrier()

    return circuit
```

```python
circuitA = sampleCircuitA(qubits=3, encode=True)
circuitA.draw(output='mpl')
```

```python
circuitA = sampleCircuitA(layer=2)
circuitA.draw(output='mpl')
```

## 1.3 Circuit C

```python
def sampleCircuitC(layer=1, qubits=4, encode=False):
    circuit = QuantumCircuit(qubits)
    
    if encode:
        for i in range(qubits):
            circuit.h(i)
            circuit.ry(np.random.uniform(0, 2 * np.pi), i)
        circuit.barrier()
    
    for i in range(layer):
        for j in range(qubits):
            circuit.ry(np.random.uniform(0, 2 * np.pi), j)
            
        circuit.crx(np.random.uniform(0, 2 * np.pi), qubits - 1, 0)
        for j in range(qubits - 1, 0, -1):
            circuit.crx(np.random.uniform(0, 2 * np.pi), j - 1, j)
        
        circuit.barrier()
        
        for j in range(qubits):
            circuit.ry(np.random.uniform(0, 2 * np.pi), j)
            
        if qubits == 3:
            circuit.crx(np.random.uniform(0, 2 * np.pi), 0, 2)
        if qubits == 4:
            circuit.crx(np.random.uniform(0, 2 * np.pi), 3, 2)
            circuit.crx(np.random.uniform(0, 2 * np.pi), 0, 3)
            circuit.crx(np.random.uniform(0, 2 * np.pi), 1, 0)
            circuit.crx(np.random.uniform(0, 2 * np.pi), 2, 1)
        if qubits == 5:
            circuit.crx(np.random.uniform(0, 2 * np.pi), 4, 3)
            circuit.crx(np.random.uniform(0, 2 * np.pi), 0, 4)
            circuit.crx(np.random.uniform(0, 2 * np.pi), 1, 0)
            circuit.crx(np.random.uniform(0, 2 * np.pi), 2, 1)
            circuit.crx(np.random.uniform(0, 2 * np.pi), 3, 2)
        
        circuit.barrier()

    return circuit
```

```python
circuitC = sampleCircuitC(qubits=5)
circuitC.draw(output='mpl')
```

```python
circuitC = sampleCircuitC(layer=2)
circuitC.draw(output='mpl')
```

## 1.4 Circuit D

```python
def sampleCircuitD(layer=1, qubits=4, encode=False):
    circuit = QuantumCircuit(qubits)
    
    if encode:
        for i in range(qubits):
            circuit.h(i)
            circuit.ry(np.random.uniform(0, 2 * np.pi), i)
        circuit.barrier()
    
    for i in range(layer):
        for j in range(qubits):
            circuit.rx(np.random.uniform(0, 2 * np.pi), j)
            circuit.rz(np.random.uniform(0, 2 * np.pi), j)
            
        for j in range(qubits - 1, -1, -1):
            for k in range(qubits - 1, -1, -1):
                if j != k:
                    circuit.crx(np.random.uniform(0, 2 * np.pi), j, k)
        circuit.barrier()
            
        for j in range(qubits):
            circuit.rx(np.random.uniform(0, 2 * np.pi), j)
            circuit.rz(np.random.uniform(0, 2 * np.pi), j)
        
        circuit.barrier()

    return circuit
```

```python
circuitD = sampleCircuitD(qubits=4)
circuitD.draw(output='mpl')
```

## 1.5 Circuit E

```python
def sampleCircuitE(layer=1, qubits=4, encode=False):
    circuit = QuantumCircuit(qubits)
    
    if encode:
        for i in range(qubits):
            circuit.h(i)
            circuit.ry(np.random.uniform(0, 2 * np.pi), i)
        circuit.barrier()
    
    for i in range(layer):
        for j in range(qubits):
            circuit.rx(np.random.uniform(0, 2 * np.pi), j)
            circuit.rz(np.random.uniform(0, 2 * np.pi), j)
            
        for j in range(1, qubits, 2):
            circuit.crx(np.random.uniform(0, 2 * np.pi), j, j - 1)
        circuit.barrier()
        
        for j in range(qubits):
            circuit.rx(np.random.uniform(0, 2 * np.pi), j)
            circuit.rz(np.random.uniform(0, 2 * np.pi), j)
        
        for j in range(2, qubits, 2):
            circuit.crx(np.random.uniform(0, 2 * np.pi), j, j - 1)
        
        circuit.barrier()

    return circuit
```

```python
circuitE = sampleCircuitE(qubits=5)
circuitE.draw(output='mpl')
```

## 1.6 Circuit F

```python
def sampleCircuitF(layer=1, qubits=4, encode=False):
    circuit = QuantumCircuit(qubits)
    
    if encode:
        for i in range(qubits):
            circuit.h(i)
            circuit.ry(np.random.uniform(0, 2 * np.pi), i)
        circuit.barrier()
    
    for i in range(layer):
        for j in range(qubits):
            circuit.ry(np.random.uniform(0, 2 * np.pi), j)
        
        for j in range(1, qubits, 2):
            circuit.cx(j - 1, j)
        for j in range(2, qubits, 2):
            circuit.cx(j - 1, j)
        
        circuit.barrier()
        
    return circuit
```

```python
circuitF = sampleCircuitF(qubits=5)
circuitF.draw(output='mpl')
```

## 1.7 Circuit G

```python
def sampleCircuitG(layer=1, qubits=4, encode=False):
    circuit = QuantumCircuit(qubits)
    
    if encode:
        for i in range(qubits):
            circuit.h(i)
            circuit.ry(np.random.uniform(0, 2 * np.pi), i)
        circuit.barrier()
    
    for i in range(layer):
        for j in range(qubits):
            circuit.rx(np.random.uniform(0, 2 * np.pi), j)
            circuit.rz(np.random.uniform(0, 2 * np.pi), j)
            
        circuit.crx(np.random.uniform(0, 2 * np.pi), qubits - 1, 0)
        for j in range(qubits - 1, 0, -1):
            circuit.crx(np.random.uniform(0, 2 * np.pi), j - 1, j)
        
        circuit.barrier()

    return circuit
```

# 2. Measure Entanglement

## 2.1 Tool functions

```python
def Q(circuit):
    matrix = getDensityMatrix(circuit)
    n = len(circuit.qubits)
    
    trs = []
    indexes = list(range(n))
    for i in range(n):
        sub_indexes = indexes.copy()
        sub_indexes.remove(i)
        
        rou = partial_trace(matrix, sub_indexes)
        tr = np.abs(np.trace(rou.data @ rou.data))
        trs.append(tr)
        
    tr = sum(trs)
    return 2 * (1 - tr / n)
```

## 2.2 Measure algorithm

```python
def Ent(sampler, *, epoch=3000, layer=1, encode=False):
    ent = 0
    for i in range(epoch):
        ent += Q(sampler(layer=layer, encode=encode))
    
    return ent / epoch
```
