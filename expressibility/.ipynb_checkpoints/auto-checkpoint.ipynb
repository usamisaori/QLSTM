```python
import numpy as np
import qiskit
from qiskit import QuantumCircuit
from qiskit.quantum_info import state_fidelity, partial_trace
```

```python
from qiskit.quantum_info import DensityMatrix, Statevector

def getDensityMatrix(circuit):
    return DensityMatrix(circuit).data

def getStatevector(circuit):
    return Statevector(circuit).data
```

```python
import matplotlib.pyplot as  plt
```

```python
def u1(circuit, index):
    circuit.u1(np.random.uniform(0, 2 * np.pi), index)

def u2(circuit, index):
    circuit.u2(np.random.uniform(0, 2 * np.pi), np.random.uniform(0, 2 * np.pi), index)
    
def u3(circuit, index):
    circuit.u3(np.random.uniform(0, 2 * np.pi), np.random.uniform(0, 2 * np.pi), np.random.uniform(0, 2 * np.pi), index)
```

# 1. Create Circuit

```python
def inputCircuit(layer=1):
    circuit = QuantumCircuit(4)
    
    for i in range(4):
        circuit.h(i)
        circuit.ry(np.random.uniform(0, 2 * np.pi), i)
    
    u3(circuit, 0)
    u1(circuit, 1)
    u2(circuit, 2)
    u1(circuit, 3)
    
    for i in range(layer):
        circuit.cz(0, 1)
        circuit.cz(1, 2)
        circuit.cz(2, 3)
        circuit.cz(3, 0)
        
        circuit.barrier()
        
        u1(circuit, 0)
        u2(circuit, 1)
        u1(circuit, 2)
        u3(circuit, 3)
    
        circuit.barrier()

    return circuit
```

```python
input_circuit = inputCircuit()
input_circuit.draw(output='mpl')
```

```python
def forgetCircuit(layer=1):
    circuit = QuantumCircuit(4)
    
    for i in range(4):
        circuit.h(i)
        circuit.ry(np.random.uniform(0, 2 * np.pi), i)
    
    u2(circuit, 0)
    u2(circuit, 1)
    u1(circuit, 2)
    u2(circuit, 3)
    
    for i in range(layer):
        circuit.cz(0, 1)
        circuit.cz(1, 2)
        circuit.cz(2, 3)
        circuit.cz(3, 0)
        
        circuit.barrier()
        
        u3(circuit, 0)
        u2(circuit, 1)
        u1(circuit, 2)
        u3(circuit, 3)
    
        circuit.barrier()

    return circuit
```

```python
forget_circuit = forgetCircuit()
forget_circuit.draw(output='mpl')
```

```python
def candidateCircuit(layer=1):
    circuit = QuantumCircuit(4)
    
    for i in range(4):
        circuit.h(i)
        circuit.ry(np.random.uniform(0, 2 * np.pi), i)
    
    u3(circuit, 0)
    u1(circuit, 1)
    u1(circuit, 2)
    u2(circuit, 3)
    
    for i in range(layer):
        circuit.cz(0, 1)
        circuit.cz(1, 2)
        circuit.cz(2, 3)
        circuit.cz(3, 0)
        
        circuit.barrier()
        
        u1(circuit, 0)
        u3(circuit, 1)
        u1(circuit, 2)
        u2(circuit, 3)
    
        circuit.barrier()

    return circuit
```

```python
candidate_circuit = candidateCircuit()
candidate_circuit.draw(output='mpl')
```

```python
def outputCircuit(layer=1):
    circuit = QuantumCircuit(4)
    
    for i in range(4):
        circuit.h(i)
        circuit.ry(np.random.uniform(0, 2 * np.pi), i)
    
    u2(circuit, 0)
    u1(circuit, 1)
    u2(circuit, 2)
    u1(circuit, 3)
    
    for i in range(layer):
        circuit.cz(0, 1)
        circuit.cz(1, 2)
        circuit.cz(2, 3)
        circuit.cz(3, 0)
        
        circuit.barrier()
        
        u1(circuit, 0)
        u1(circuit, 1)
        u3(circuit, 2)
        u1(circuit, 3)
        
        circuit.barrier()

    return circuit
```

```python
output_circuit = outputCircuit()
output_circuit.draw(output='mpl')
```

# 2. Measure Expressibililty

```python
def P_haar(N, F):
    if F == 1:
        return 0
    return (N - 1) * ((1 - F) ** (N - 2))
```

```python
def KL(P, Q):
    epsilon = 1e-8
    kl_divergence = 0.0
    
    for p, q in zip(P, Q):
        kl_divergence += p * np.log( (p + epsilon) / (q + epsilon) )
    
    return abs(kl_divergence)
```

```python
def expressibility(sampler, *, bins=100, epoch=3000, layer=1):
    unit = 1 / bins
    limits = []
    probabilities = np.array([0] * bins)
    for i in range(1, bins + 1):
        limits.append(unit * i)

    for i in range(epoch):
        circuit_1 = sampler(layer=layer)
        circuit_2 = sampler(layer=layer)
        f = state_fidelity(
            getStatevector(circuit_1),
            getStatevector(circuit_2)
        )

        for j in range(bins):
            if f <= limits[j]:
                probabilities[j] += 1
                break

    pHaar_vqc = [ P_haar(2 ** 4, f - (unit/2)) / bins for f in limits]
    probabilities = [ p / epoch for p in probabilities ]

    return -np.log10(KL(probabilities, pHaar_vqc))
```

# 3. Measure Entanglement

```python
def Q(circuit):
    matrix = getDensityMatrix(circuit)
    
    rou0 = partial_trace(matrix, [1, 2, 3])
    rou1 = partial_trace(matrix, [0, 2, 3])
    rou2 = partial_trace(matrix, [0, 1, 3])
    rou3 = partial_trace(matrix, [0, 1, 2])
    
    tr0 = np.abs(np.trace(rou0.data @ rou0.data))
    tr1 = np.abs(np.trace(rou1.data @ rou1.data))
    tr2 = np.abs(np.trace(rou2.data @ rou2.data))
    tr3 = np.abs(np.trace(rou3.data @ rou3.data))
    
    tr = tr0 + tr1 + tr2 + tr3
    
    return 2 * (1 - tr / 4)
```

```python
def Ent(sampler, *, epoch=5000, layer=1):
    ent = 0
    for i in range(epoch):
        ent += Q(sampler(layer=layer))
    
    return ent / epoch
```
