# 1. Circuits Sampler

```python
theta = Parameter("θ")
phi = Parameter("φ")
lamb = Parameter("λ")
```

## 1.1 Circuit A

```python
def sampleCircuitA(layer=1, qubits=4):
    circuit = QuantumCircuit(qubits)
    
    for i in range(layer):
        for i in range(qubits - 1):
            circuit.cx(i, i + 1)
        circuit.cx(qubits - 1, 0)
        
        for i in range(qubits - 1):
            circuit.cx(i, i + 1)
        circuit.cx(qubits - 1, 0)
        
        circuit.barrier()
        
        for i in range(qubits):
            circuit.u3(theta, phi, lamb, i)

    return circuit
```

```python
circuitA = sampleCircuitA(qubits=4)
circuitA.draw(output='mpl')
```

## 1.2 Circuit B

```python
def sampleCircuitB1(layer=1, qubits=4):
    circuit = QuantumCircuit(qubits)
    
    for i in range(qubits):
        circuit.u1(theta, i)
    
    for i in range(layer):
        for j in range(qubits - 1):
            circuit.cz(j, j + 1)
        circuit.cz(qubits - 1, 0)

        circuit.barrier()
        
        for j in range(qubits):
            circuit.u1(theta, j)

    return circuit
```

```python
def sampleCircuitB2(layer=1, qubits=4):
    circuit = QuantumCircuit(qubits)
    
    for i in range(qubits):
        circuit.u2(phi, lamb, i)
    
    for i in range(layer):
        for j in range(qubits - 1):
            circuit.cz(j, j + 1)
        circuit.cz(qubits - 1, 0)
        
        circuit.barrier()
        
        
        for j in range(qubits):
            circuit.u2(phi, lamb, j)

    return circuit
```

```python
def sampleCircuitB3(layer=1, qubits=4):
    circuit = QuantumCircuit(qubits)
    
    for i in range(qubits):
        circuit.u3(theta, phi, lamb, i)
    
    for i in range(layer):
        for j in range(qubits - 1):
            circuit.cz(j, j + 1)
        circuit.cz(qubits - 1, 0)
        
        circuit.barrier()
        
        for j in range(qubits):
            circuit.u3(theta, phi, lamb, j)
        
    return circuit
```

```python
circuitB1 = sampleCircuitB1(qubits=4)
circuitB1.draw(output='mpl')
```

```python
circuitB2 = sampleCircuitB2(qubits=4)
circuitB2.draw(output='mpl')
```

```python
circuitB3 = sampleCircuitB3(qubits=4)
circuitB3.draw(output='mpl')
```

## 1.3 Circuit C

```python
def sampleCircuitC(layer=1, qubits=4):
    circuit = QuantumCircuit(qubits)
    
    for i in range(layer):
        for j in range(qubits):
            circuit.ry(theta, j)
            
        circuit.crx(theta, qubits - 1, 0)
        for j in range(qubits - 1, 0, -1):
            circuit.crx(theta, j - 1, j)
        
        circuit.barrier()
        
        for j in range(qubits):
            circuit.ry(theta, j)
            
        circuit.crx(theta, 3, 2)
        circuit.crx(theta, 0, 3)
        circuit.crx(theta, 1, 0)
        circuit.crx(theta, 2, 1)
        
    return circuit
```

```python
circuitC = sampleCircuitC(qubits=4)
circuitC.draw(output='mpl')
```

## 1.4 Circuit D

```python
def sampleCircuitD(layer=1, qubits=4):
    circuit = QuantumCircuit(qubits)
    
    for i in range(layer):
        for j in range(qubits):
            circuit.rx(theta, j)
            circuit.rz(theta, j)
            
        for j in range(qubits - 1, -1, -1):
            for k in range(qubits - 1, -1, -1):
                if j != k:
                    circuit.crx(theta, j, k)
        circuit.barrier()
            
        for j in range(qubits):
            circuit.rx(theta, j)
            circuit.rz(theta, j)
        
    return circuit
```

```python
circuitD = sampleCircuitD(qubits=4)
circuitD.draw(output='mpl')
```

## 1.5 Circuit E

```python
def sampleCircuitE(layer=1, qubits=4):
    circuit = QuantumCircuit(qubits)

    for i in range(layer):
        for j in range(qubits):
            circuit.rx(theta, j)
            circuit.rz(theta, j)
            
        for j in range(1, qubits, 2):
            circuit.crx(theta, j, j - 1)
        
        for j in range(qubits):
            circuit.rx(theta, j)
            circuit.rz(theta, j)
        
        for j in range(2, qubits, 2):
            circuit.crx(theta, j, j - 1)
        
    return circuit
```

```python
circuitE = sampleCircuitE(qubits=4)
circuitE.draw(output='mpl')
```

## 1.6 Circuit G

```python
def sampleCircuitG(layer=1, qubits=4):
    circuit = QuantumCircuit(qubits)
    
    for i in range(layer):
        for j in range(qubits):
            circuit.rx(theta, j)
            circuit.rz(theta, j)
            
        circuit.crx(theta, qubits - 1, 0)
        for j in range(qubits - 1, 0, -1):
            circuit.crx(theta, j - 1, j)

    return circuit
```

```python
circuitG = sampleCircuitG(qubits=4)
circuitG.draw(output='mpl')
```

## Encoding Circuit

```python
def sampleEncoding(qubits):
    circuit = QuantumCircuit(qubits)
    
    for i in range(qubits):
        circuit.h(i)
        circuit.ry(theta, i)
        
    return circuit
```

```python
circuit = sampleEncoding(4)
circuit.draw(output='mpl')
```

```python
circuit = sampleEncoding(5).compose(sampleCircuitB3(layer=2, qubits=5))
circuit.draw(output='mpl')
```

# 2. Auto circuits

```python
def autoInputCircuit():
    circuit = QuantumCircuit(4)
    
    circuit.u3(theta, phi, lamb, 0)
    circuit.u1(theta, 1)
    circuit.u2(phi, lamb, 2)
    circuit.u1(theta, 3)
    
    for i in range(4 - 1):
        circuit.cz(i, i + 1)
    circuit.cz(3, 0)
    
    circuit.barrier()
    
    circuit.u1(theta, 0)
    circuit.u2(phi, lamb, 1)
    circuit.u1(theta, 2)
    circuit.u3(theta, phi, lamb, 3)
    
    return circuit
```

```python
inputCircuit = autoInputCircuit()
inputCircuit.draw(output='mpl')
```

```python
def autoForgetCircuit():
    circuit = QuantumCircuit(4)
    
    circuit.u2(phi, lamb, 0)
    circuit.u2(phi, lamb, 1)
    circuit.u1(theta, 2)
    circuit.u2(phi, lamb, 3)
    
    for i in range(4 - 1):
        circuit.cz(i, i + 1)
    circuit.cz(3, 0)
    
    circuit.barrier()
    
    circuit.u3(theta, phi, lamb, 0)
    circuit.u2(phi, lamb, 1)
    circuit.u1(theta, 2)
    circuit.u3(theta, phi, lamb, 3)
    
    return circuit
```

```python
forgetCircuit = autoForgetCircuit()
forgetCircuit.draw(output='mpl')
```

```python
def autoCandidateCircuit():
    circuit = QuantumCircuit(4)
    
    circuit.u3(theta, phi, lamb, 0)
    circuit.u1(theta, 1)
    circuit.u1(theta, 2)
    circuit.u2(phi, lamb, 3)
    
    for i in range(4 - 1):
        circuit.cz(i, i + 1)
    circuit.cz(3, 0)
    
    circuit.barrier()
    
    circuit.u1(theta, 0)
    circuit.u3(theta, phi, lamb, 1)
    circuit.u1(theta, 2)
    circuit.u2(phi, lamb, 3)
    
    return circuit
```

```python
candidateCircuit = autoCandidateCircuit()
candidateCircuit.draw(output="mpl")
```
